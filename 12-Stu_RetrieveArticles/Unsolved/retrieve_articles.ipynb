```python
# Dependencies
import requests
from config import api_key
import time

url = "https://api.nytimes.com/svc/search/v2/articlesearch.json?"

# Store a search term
query = "election"

# Search for articles published between a begin and end date
## why does this give different results than fq using pub year? it should be the same results
## perhaps they're sorted differently due to result prioritization?
#filter = f"begin_date=20140101&end_date=20141231"
filter = f"fq=pub_year:(2014)"

# Build URL
query_url = f"{url}api-key={api_key}&q={query}&{filter}"
```

```python
articles = requests.get(query_url).json()
articles
```

```python
# Retrieve articles
articles_list = articles['response']['docs']
for i in range(10):
    print(f"A snippet from the article: {articles_list[i]['snippet']}\n---------------------------")
```

```python
# BONUS: How would we get 30 results? 
# HINT: Look up the page query param
## well, this gets 30 results. the README says i should get a 429: too many requests error
## but it doesn't seem to get that. it just gives 30 results, which is what the task was, i guess
error_articles = []
for i in range(3):
    error_query = f"{query_url}&page={i}"
    response_articles = requests.get(error_query).json()["response"]["docs"]
    error_articles.extend(response_articles)
error_articles
```
